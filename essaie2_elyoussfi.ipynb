## Vectorisation

```python
from sklearn.feature_extraction.text import TfidfVectorizer

transformer = TfidfVectorizer().fit(nettoyer_description_train)
```

```python
print("NB features: " + str((len(transformer.vocabulary_))), "\n")
print("type :\n", type(transformer.vocabulary_), "\n")
transformer.vocabulary_ #print le dict
```

# X_train

```python
X_train = transformer.transform(nettoyer_description_train)
print("shape :\n", np.shape(X_train), "\n")
print("type :\n", type(X_train), "\n")
X_train
```

# X_test

```python
X_test = transformer.transform(nettoyer_description_test)
print("shape :\n", np.shape(X_test), "\n")
print("type :\n", type(X_test), "\n")
X_test
```

# y_train

```python
y_train = df_y_train["Category"].values
print("shape :\n", np.shape(y_train), "\n")
print("type :\n", type(y_train), "\n")
y_train
```

# Modele

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter = 200)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print("shape :\n", np.shape(predictions), "\n")
print("type :\n", type(predictions), "\n")
predictions
```

```python
df_X_test["Category"] = predictions
df_X_test
```
